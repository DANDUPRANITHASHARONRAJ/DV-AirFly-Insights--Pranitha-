```python
#CODE FOR GROUPING COLS BY TYPE(OBJ/NUM)
def group_columns_by_obj_or_float(df):
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    print("Numeric columns (int/float) are:")
    if numeric_cols:
        for col in numeric_cols:
            print(f"  - {col}")
    else:
        print("none found")
    
    print("\n Categorical columns (object/category are):")
    if categorical_cols:
        for col in categorical_cols:
            print(f"  - {col}")
    else:
        print("none found")
    
    return numeric_cols, categorical_cols

```

<span style="color:yellow; font-size:20px;">DATE FORMAT SET TO HH:MM</span>

```python
#CHANGING 6 COLS DATA FROM HHMM TO HH:MM
import pandas as pd

# Columns to convert

time_cols = ['CRS_DEP_TIME', 'DEP_TIME', 'WHEELS_OFF',
             'WHEELS_ON', 'CRS_ARR_TIME', 'ARR_TIME']

# Function to convert custom HHMM to HH:MM
def convert_custom_hhmm(value):
    if pd.isna(value):
        return ""
    try:
        val = int(value)
        # For values < 100, treat as 0th hour
        hour = val // 100 if val >= 100 else 0
        minute = val % 100 if val >= 100 else val
        return f"{hour:02d}:{minute:02d}"
    except:
        return ""

# Apply conversion for all time columns
for col in time_cols:
    df[col + '_HHMM'] = df[col].apply(convert_custom_hhmm)

# Check first 20 rows
print(df.head(20))
```

```python
#DROPPING ORIGINAL HHMM
# Drop the original HHMM columns
df = df.drop(columns=['CRS_DEP_TIME', 'DEP_TIME', 'WHEELS_OFF',  'WHEELS_ON', 'CRS_ARR_TIME', 'ARR_TIME'])
```

```python
#RENAMING NEW HH:MM COLS WITH OLD HHMM NAMES
df = df.rename(columns={
    'CRS_DEP_TIME_HHMM': 'CRS_DEP_TIME',
    'DEP_TIME_HHMM': 'DEP_TIME',
    'WHEELS_OFF_HHMM': 'WHEELS_OFF',
    'WHEELS_ON_HHMM': 'WHEELS_ON',
    'CRS_ARR_TIME_HHMM': 'CRS_ARR_TIME',
    'ARR_TIME_HHMM': 'ARR_TIME'
})
```

```python
#ROUTE
# Create Route as ORIGIN-DEST
df['Route'] = df['ORIGIN'] + '-' + df['DEST']

# Check the first few rows
print(df[['ORIGIN', 'DEST', 'Route']].head())
```

```python
#AVG DELAY BY AIRLINE
airline_delays = df.groupby('AIRLINE')[[
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]].mean().reset_index()

airline_delays

```

```python
#AVG DELAY BY AIRLINE,DELAY TYPE
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
airline_delays_melted = airline_delays.melt(id_vars='AIRLINE', var_name='Delay_Type', value_name='Avg_Delay')
sns.barplot(data=airline_delays_melted, x='AIRLINE', y='Avg_Delay', hue='Delay_Type')
plt.title("Average Delay by Airline and Delay Type")
plt.ylabel("Average Delay (minutes)")
plt.xticks(rotation=90)
plt.show()
```

```python
#AVG OF DELAY_TYPES
delay_types = ['DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER', 'DELAY_DUE_NAS','DELAY_DUE_SECURITY','DELAY_DUE_LATE_AIRCRAFT']
df[delay_types].mean()
```

```python
# Calculate average delay per type
avg_delays = df[delay_types].mean().reset_index()
avg_delays.columns = ['Delay_Type', 'Avg_Delay']

# Plot as bar chart
plt.figure(figsize=(8,6))
sns.barplot(data=avg_delays, x='Delay_Type', y='Avg_Delay', palette="viridis")
plt.title("Average Delay by Type")
plt.ylabel("Average Delay (minutes)")
plt.xlabel("Delay Type")
plt.xticks(rotation=45)
plt.show()
```

```python
#AVG DELAY WITH TIME
bins = [0,6,12,18,24]
labels = ['Night','Morning','Afternoon','Evening']
df['DEP_PERIOD'] = pd.cut(df['DEP_TIME_HOUR'], bins=bins, labels=labels, right=False)

# Average delay by period
delay_by_period = df.groupby('DEP_PERIOD')[delay_types].mean().reset_index()

sns.barplot(data=delay_by_period.melt(id_vars='DEP_PERIOD'), x='DEP_PERIOD', y='value', hue='variable')
plt.title("Average Delay by Time of Day")
plt.ylabel("Average Delay (minutes)")
plt.show()
```

```python
#TOP 10 ORIGIN AIRPORTS BY TOTAL AVG DELAY
airport_delays = df.groupby('ORIGIN')[delay_types].mean().sum(axis=1).sort_values(ascending=False).head(10)

airport_delays.plot(kind='bar', figsize=(12,6))
plt.title("Top 10 Origin Airports by Total Average Delay")
plt.ylabel("Total Average Delay (minutes)")
plt.show()
```

```python
#TOP 20 BUSIEST AIRPORTS,AV3ERAGE ARRIVAL DELAYS
import matplotlib.pyplot as plt
import seaborn as sns

# Top 20 airports by number of flights
airport_counts = df['ORIGIN'].value_counts().head(20).reset_index()
airport_counts.columns = ['IATA', 'flights']

# Average delay per airport
avg_delay = df.groupby('ORIGIN')['ARR_DELAY'].mean().reset_index()
avg_delay.columns = ['IATA', 'avg_delay']

# Merge
summary = airport_counts.merge(avg_delay, on='IATA')

# Plot
plt.figure(figsize=(12,6))
sns.barplot(data=summary, x='flights', y='IATA', hue='avg_delay', dodge=False, palette="coolwarm")
plt.title("Top 20 Busiest Airports and Their Average Arrival Delays")
plt.xlabel("Number of Flights")
plt.ylabel("Airport")
plt.legend(title="Avg Delay (min)")
plt.show()
```

```python
#AVERAGE ARRIVAL DELAY BY SEASON
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define season map
season_map = {
    12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Fall', 10: 'Fall', 11: 'Fall'
}

# Group by season without creating a new col
season_delay = df.groupby(df['MONTH'].map(season_map))['ARR_DELAY'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=season_delay,
            x='MONTH', y='ARR_DELAY',
            palette="magma",
            order=['Winter','Spring','Summer','Fall'])
plt.title("Average Arrival Delay by Season")
plt.xlabel("Season")
plt.ylabel("Avg Delay (minutes)")
plt.show()
```
